This notebook generates the plot in the top right of Figure 1, which shows a portion of the methane IR spectrum.

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot figure
ir_spectra = np.load('/content/10_IR_Spectra_date_7_5.npy')
methane_x_axis = ir_spectra[0]
methane_y_axis = (ir_spectra[1]) / (np.max(ir_spectra[1]))

figure = plt.figure(figsize = (8, 3))
axes = figure.add_subplot(1, 1, 1)
#axes.set_ylabel("Normalized Transmittance", fontsize = 16)
axes.tick_params(labelsize = 25)
#axes.invert_xaxis()

axes.plot(methane_x_axis, methane_y_axis, color = 'brown')
axes.set_xlim(1245, 1355)
```
